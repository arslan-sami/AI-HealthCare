```python
if Selection == 'symptoms' or Selection == 'Symptoms' or Selection == 'Symptom' or Selection == 'symptom' :
  x = np.array(Disease_symptoms['Symptoms'])
  y = Disease_symptoms['Precautions']
  
  cv = CountVectorizer(
    stop_words='english',
    #ngram_range=(1, 1),  #ngram_range=(1, 1) is the default
    )
  
  X = cv.fit_transform(x)
  svc_s = SVC()

  X_train , X_test , Y_train , Y_test = train_test_split(X , y , test_size=0.2 )
  svc_s.fit(X_train , Y_train)

  problem1 = input("Enter The Symptom 1 : ")
  problem2 = input("Enter The Symptom 2 : ")
  problem3 = input("Enter The Symptom 3 : ")
  problem4 = input("Enter The Symptom 4 : ")
  
#   problem1 = problem1.replace(" ","")
#   problem2 = problem2.replace(" ","")
#   problem3 = problem3.replace(" ","")
#   problem4 = problem4.replace(" ","")

  problem = problem1 + "," + problem2 + "," + problem3 + "," + problem4
  print(problem)
  vectors = cv.transform([problem]).toarray()

elif Selection == 'Disease' or Selection == 'disease' or Selection == 'Diseases' or Selection == 'diseases':
  x = Disease_symptoms['Disease']
  y = Disease_symptoms['Precautions']
  cv = CountVectorizer(
    stop_words='english',
    ngram_range=(1, 1),  #ngram_range=(1, 1) is the default
    dtype='double')
  X = cv.fit_transform(x)
  svc_d = SVC()
  X_train,X_test,Y_train, Y_test = train_test_split(X,y, test_size = 0.2 , random_state=101)
  svc_d.fit(X_train,Y_train)
  problems = input("Tell me your Problem! : ")
  vectors = cv.transform([problems]).toarray()
# muscleweakness,stiffneck,swellingjoints,painfulwalking	
```
